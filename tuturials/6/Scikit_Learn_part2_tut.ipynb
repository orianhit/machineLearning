#### Dataset: The Iris dataset

This dataset was introduced in 1936 by the statistician [Sir Ronald A. Fisher](https://en.wikipedia.org/wiki/Ronald_Fisher). 

The dataset contains 4 features (attributes) of 50 samples containing 3 different species of iris plants. The goal is to classify the species of iris plant given the attributes. Sklearn already includes a subset of this data, we will use sklearn's `load_iris` function to load the data.



**Classes:**
+ Iris Setosa 
- Iris Versicolour 
- Iris Virginica

**Features (attributes):**
+ sepal length (cm) 
- sepal width (cm) 
- petal length (cm) 
- petal width (cm) 


Let's start with the relevant imports. Pay attention to the `KNeighborsClassifier` class for the KNN algorithm. You can also learn more using the online documentation for [knn](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

```python
# Importing implementation of KNN
from sklearn.neighbors import KNeighborsClassifier

# Importing module responsible for performance evaluation
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

# Importing train-test data-set splitting functionality
from sklearn.model_selection import train_test_split
```

#### Dataset visualization

Let's visualize the data and see what we got.. As mentioned, for visulazation reasons we will use only the first 2 columns (2D is easier to visualize than 4D), and so we will do in the rest of this notebook.

```python
# For easy plotting and interpretation, we only use first 2 features here. 
# We're throwing away useful information - don't do this at home! 
X = iris.data[:,:2]  
y = iris.target

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# plot data
plt.figure(figsize=(10,8))
plt.scatter(X[:, 0], X[:, 1], c=y,  marker="o", cmap=cmap_bold, s=200)


x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)    
plt.title('Iris dataset')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

#### Iris dataset classification using KNN

As you can see, it would be fairly easy to separate the "red" irises from the two classes. However, separating the "green" and "blue" ones would be more challenging. 

Let's use KNN to do it. 

But, first let's split the data to train and test

```python
# Split into training and test sets - returned data represented as 'numpy.ndarray'

XTrain, XTest, yTrain, yTest = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
print(XTrain[:10])
print(yTrain[:10])
```

```python
print(XTrain.shape)
print(yTrain.shape)
```

```python
print(XTest.shape)
print(yTest.shape)
```

Now we will train the classifier on the train set. As always in sklearn based classifiers, we do it using the **`fit`** function. Key question is what $K$ to chose. Let's start with **3**

```python
# set up the model, k-NN classification with k = ?  
k = 3
clf = KNeighborsClassifier(n_neighbors=k)
clf.fit(XTrain, yTrain)
```

Let's see how good it is!
For prediction we will use sklearn's **`predict`** function and apply it on the X matrix of the **test** set **XTest** against the **ground truth** test labels **yTest**. We will then asses it's performance by printing the confusion matrix, followed by the accuracy score.

```python
y_pred=clf.predict(XTest)

print(metrics.confusion_matrix(y_true = yTest, y_pred = y_pred))
print('Accuracy = ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
print(classification_report(yTest,y_pred,digits=4))
```

### Cross validation 

The simplest way to use cross-validation is to use the `cross_val_score` function sklearn provide you with.

The following example demonstrates how to estimate the accuracy of a KNN model on the iris dataset by splitting the data, fitting a model and computing the score 10 consecutive times (with different splits each time). Note that the function needs to get as parameters the classifier, the feature matrix (X), the labels (y), and a parameter that indicates how many folds to execute (cv). The function returns a list with the scores for each fold.

```python
from sklearn.model_selection import cross_val_score

clf = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(clf, X, y, cv=10)

scores
```

This gave us a list with the accuracy of each iteration. If you want to get the overall accuracy, you can simply write:

```python
print(f"Accuracy: {scores.mean()}")
```

#### The Naive Bayes algorithm
Naive Bayes models are a group of extremely fast and simple classification algorithms that are often suitable for very high-dimensional datasets. Because they are so fast and have so few tunable parameters, they end up being very useful as a quick-and-dirty baseline for a classification problem. 

The Naive Bayes classifier aggregates information using conditional probability with an assumption of independence among features. What does it mean? For example, it means we have to assume that the comfort of the room on the Titanic is independent of the fare ticket. This assumption is absolutely wrong and it is why it is called Naive. It allows to simplify the calculation, even on very large datasets.

Let's implement a classic Gaussian Naive Bayes on the Titanic Disaster dataset. We will use the same features as in previous example to predict survival.

```python
# Import sklearn implementation of Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
# Prepare Data
X, y = datasets.load_iris(return_X_y=True)
XTrain, XTest, yTrain, yTest = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Instantiate the classifier
gnb = GaussianNB()


# Train classifier
gnb.fit(XTrain, yTrain)
```

```python
# Use the trained classifer to predit 
y_pred = gnb.predict(XTest)
y_pred_train = gnb.predict(XTrain)

# Evaluate Accuracy
print('Accuracy on Train data=', metrics.accuracy_score(y_true=yTrain, y_pred=y_pred_train))
print('Accuracy on test data=', metrics.accuracy_score(y_true=yTest, y_pred=y_pred))
```

#### Our first tree

Before we start building our trees , we will define an helper functions:

`split_data` - this functions gets a list of attributes, and returns a train test split, where only the listed attributes appear in the $X$ matrix

```python
def split_data(titanic_df, features):
    """Split a subset of the titanic dataset, given by the features, into train and test sets."""
    titanic_predictors = titanic_df[features].values
    titanic_labels = titanic_df["Survived"].values

    # Split into training and test sets
    XTrain, XTest, yTrain, yTest = train_test_split(titanic_predictors, titanic_labels, random_state=42, test_size=0.2)
    return XTrain, XTest, yTrain, yTest
```

After we are done with the prep, let's start building out first, yet very simple tree - one that **ONLY operates on sex**! 

```python
# Use only "Sex" column and split data to train and test
XTrain, XTest, yTrain, yTest = split_data(titanic_df, ["Sex"])
```

```python
# Create Decision Tree Classifier instance
decisionTree = tree.DecisionTreeClassifier()

# fit the tree with the traing data
decisionTree = decisionTree.fit(XTrain, yTrain)
```

```python
# predict with the training data
y_pred_train = decisionTree.predict(XTrain)
# measure accuracy
print('Accuracy on training data =', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))


# predict with the test data
y_pred = decisionTree.predict(XTest)
# measure accuracy
print('Accuracy on test data =', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

~78% correct on the test set isn't bad! - sex seems to be a very good indicator of whether someone has survived or not.

You can see that we started with 445 samples (based on the value of "samples" in the top box), out of which 159 survived and 286 died (based on the numbers "value" of the same box). Therfore at the root, the majority is "perished" (value of "class"). If you make decision point, according to "sex" (the top value in this box), you see that female (sex<0.5) we have 157 samples out of them 166 survived (therefore class is "survived"), while for male (sex not smaller than 0.5) only 43 out pf 288 survived - therefore class is perished

Let's add more features.

```python
# Use only "Sex" and "SibSp" columns and split data to train and test
used_features = ["Sex", "SibSp"]
XTrain, XTest, yTrain, yTest = split_data(titanic_df, used_features)

decisionTree = tree.DecisionTreeClassifier()
decisionTree = decisionTree.fit(XTrain, yTrain)

y_pred_train = decisionTree.predict(XTrain)
print('Accuracy on training data= ', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))

y_pred = decisionTree.predict(XTest)
print('Accuracy on test data= ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

Our accuracy on the training data has and test set has gone a bit up, let's see what happens when we use more features

```python
all_features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]

XTrain, XTest, yTrain, yTest = split_data(titanic_df, all_features)
decisionTree = tree.DecisionTreeClassifier()
decisionTree = decisionTree.fit(XTrain, yTrain)

y_pred_train = decisionTree.predict(XTrain)
print('Accuracy on training data= ', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))

y_pred = decisionTree.predict(XTest)
print('Accuracy on test data= ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

OK, clearly, we're overfitting the data - 98% accuracy on the training data and only ~78% on the test data. 

#### Handling overfitting - pre-pruning


As we learned Decision trees have a tendency to overfit the data. There are two parameters that help us reign in overfitting:

* **max_depth:** The maximum depth of the tree. If this is not set, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
* **min_samples_split:** The minimum number of samples required to split an internal node: If the value is an integer, then consider `min_samples_split` as the minimum number. If it is float, then `min_samples_split` is a percentage and `ceil(min_samples_split * n_samples)` are the minimum number of samples for each split.

Let's try a combination of depth and min split size (feel free to try additional combinations and see what you get, both in terms of tree complexity as well as results accuracy):

```python
decisionTree = tree.DecisionTreeClassifier(max_depth=2, min_samples_split=20)

decisionTree = decisionTree.fit(XTrain, yTrain)

y_pred_train = decisionTree.predict(XTrain)
print('Accuracy on training data= ', metrics.accuracy_score(y_true = yTrain, y_pred = y_pred_train))

y_pred = decisionTree.predict(XTest)
print('Accuracy on test data= ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

### Hyper parameter tuning:

We can use the `sklearn.model_selection`'s function `GridSearchCV` in order to perform hyper parameter tuning.<br/>
This function performs a grid search on k-fold cross validation.<br/>
This means that for each predefined hyper-parameter and each predefined value it finds the best combination of hyper-parameter, value pairs for a given predefined evaluation metric. <br/>

For more information visit the [GridSearchCV documentation](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html)

The following is an example of the usage of this function, for a KNN classifier

```python
from sklearn.model_selection import GridSearchCV 

knn_classifier = KNeighborsClassifier()
knn_params = {
      'n_neighbors': [3, 5, 7, 9],
      'metric': ['euclidean', 'manhattan']
}

cls_cv = GridSearchCV(knn_classifier,knn_params,cv=10)

grid_search_res = cls_cv.fit(X, y)
'best_params:'
grid_search_res.best_params_
'best accuracy score:'
grid_search_res.best_score_
'average mean(cross-validation) accuracy:'
grid_search_res.cv_results_['mean_test_score'].mean()
'average std(cross-validation) accuracy:'
grid_search_res.cv_results_['std_test_score'].mean()
```
